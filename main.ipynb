```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib
```

# US Data
subset of the Data only looking at US Data for youtube videos 

```python
us_data = pd.read_csv('data/USvideos.csv')
us_data.head()
```

```python
us_data.columns
#List of columns in the data frame
```

```python
us_data_cut = ['views','likes','dislikes','comment_count']
c_data = us_data[us_data_cut].corr()
c_data
```

Correlation between views likes dislikes and comment count.

The above shows the pearson correlation coefficient it works as follows:

        1 = a strong positive correlation
        0 = no correlation
       -1 = strong negative correlation.

```python
# Heat Map for the correlation 
matplotlib.pyplot.figure(figsize=(8,6))
sns.heatmap(c_data, annot=True);
```

## Cleaned Heatmap that is easier to read 

```python
fig, ax = matplotlib.pyplot.subplots(figsize=(8,6))
mask = np.zeros_like(c_data)
mask[np.triu_indices_from(mask)] = 1
sns.heatmap(c_data, mask= mask, ax= ax, annot= True);
```

## Scatterplot Diagram for views vs likes (USA)

```python
matplotlib.pyplot.figure(figsize=(16,16))
sns.scatterplot(x=us_data['views'], y=us_data['likes']);
```

# Finding the Correlation with the rest of the Data

```python
cut = ['views','likes','dislikes','comment_count']

us_data = pd.read_csv('data/USvideos.csv',usecols=cut) #Data for US
ca_data = pd.read_csv('data/CAvideos.csv',usecols=cut) #Data for Canada
de_data = pd.read_csv('data/DEvideos.csv',usecols=cut) #Data for Denmark
fr_data = pd.read_csv('data/FRvideos.csv',usecols=cut) #Data for France
gb_data = pd.read_csv('data/GBvideos.csv',usecols=cut) #Data for Great Britain
in_data = pd.read_csv('data/INvideos.csv',usecols=cut) #Data for India
jp_data = pd.read_csv('data/JPvideos.csv',usecols=cut) #Data for japan
kr_data = pd.read_csv('data/KRvideos.csv',usecols=cut) #Data for Korea 
mx_data = pd.read_csv('data/MXvideos.csv',usecols=cut) #Data for Mexico
ru_data = pd.read_csv('data/RUvideos.csv',usecols=cut) #Data for Russia

#data frame list of all of the dataframes
df_list = [us_data, ca_data, de_data, fr_data, gb_data, in_data, jp_data, kr_data, mx_data, ru_data]
#List of names
df_names = ['USA','Canada','Denmark','France','GB','India','Japan','Korea','Mexico','Russia']



views = pd.DataFrame(columns=['views', 'likes', 'dislikes', 'comment_count'])
likes = pd.DataFrame(columns=['views', 'likes', 'dislikes', 'comment_count'])
```

```python
# Correlation for each country

count = 0
while count != 10:
    # print(df_names[count])
    current_df = df_list[count]
    _x = current_df[cut].corr()
    # display(_x)
    views.loc[count] = _x.loc['views']
    likes.loc[count] = _x.loc['likes']
    count += 1
    
```

```python
views = views.drop(axis=1, labels='views') #remove redundant column
views.index = df_names
display(views.style.background_gradient())
```

```python
main_df = pd.DataFrame(columns=['views', 'likes', 'dislikes', 'comment_count', 'country'])

count = 0
entries = 0
while count != 10:
    current_df = df_list[count]
    entries = entries + len(current_df)
    country_name = df_names[count]
    current_df['country'] = country_name
    main_df = pd.merge(main_df, current_df, how='outer')
    count += 1
print(entries)
```

```python
# Correlation plot for all of the countries
main_df[cut].corr().style.background_gradient()
```

# Scatterplot for correlation of all countries

```python
matplotlib.pyplot.figure(figsize=(16,16))
sns.scatterplot(x=main_df["views"], y=main_df["likes"], hue=main_df["country"])
sns.despine(bottom=True, left=True);
```
